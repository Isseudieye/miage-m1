##### L'idée est de refaire le classement de la PL 2021 en retrouvant les équipes en haut du tableau

Créeons une fonction qui permet de calculer le nombre de point pour chaque

```python
def compute_nb_point(row):
    if row["ftr"] == "H":
        row["fthp"] = 3
        row["ftap"] = 0
    if row["ftr"] == "A":
        row["ftap"] = 3
        row["fthp"] = 0
    if row["ftr"] == "D":
        row["ftap"] = 1
        row["fthp"] = 1
    return row

# chelsea 2 - 1 arsenal #fthp 3 ftap 0
# chelsea 1 - 1 arsenal #fthp 1 ftap 1
# chelsea 1 - 2 arsenal #fthp 0 ftap 3
```

##### Idée: calculer d'abord les points des équipes: pd apply groupby 

```python
#TODO
pl_df = pl_df.apply(compute_nb_point, axis=1)
```

```python
data_pl = pl_df[["hometeam", "awayteam", "fthp", "ftap", "matchdate"]]
```

```python
data_pl
```

```python
home_result_df = data_pl.groupby("hometeam").agg({"fthp": "sum"}).reset_index().rename(columns={"hometeam": "team"})
```

```python
away_result_df = data_pl.groupby("awayteam").agg({"ftap": "sum"}).reset_index().rename(columns={"awayteam": "team"})
```

```python
home_result_df.head()
```

```python
away_result_df.head()
```

```python
pl_result = pd.merge(left=home_result_df, right=away_result_df, on="team")
```

```python
pl_result.head()
```

```python
pl_result["pts"] = pl_result["fthp"] + pl_result["ftap"]
```

```python
pl_result = pl_result[["team", "pts"]].sort_values(by="pts", ascending=False)
```

```python
pl_result.head()
```

```python
pl_result = pl_result.reset_index(drop=True)
```
